```python
import numpy as np
```

```python
# build up the dictionary of OTU,district -> mean abundance
meanMap={}
otus=[]
with open('mean.txt') as f:
    lines=f.readlines()
    header=lines[0].replace('_mean','').split()
    del header[0]
    del lines[0]
    for line in lines:
        row=line.split()
        otuName=row[0].replace('.0','')
        otus.append(otuName)
        del row[0]
        for i in range(len(row)):
            if header[i] in meanMap:
                meanMap[header[i]][otuName]=float(row[i])
            else:
                meanMap[header[i]]={}
                meanMap[header[i]][otuName]=float(row[i])
print(meanMap)
```

```python
# build up the dictionary of samples count for each district
count={}
with open('count.txt') as f:
    for line in f:
        dist,countNum=line.split()
        count[dist]=int(countNum)
print(count)
```

```python
# dictionary for district -> list of sampleID
sampleDist={}
samples=[]
with open('sample.txt') as f:
    for line in f:
        sampleID,dist=line.split()
        samples.append(sampleID)
        if dist in sampleDist:
            sampleDist[dist].append(sampleID)
        else:
            sampleDist[dist]=[sampleID]
        
print(sampleDist)
```

```python
import pandas as pd
```

```python
# build dataframe of samplesID as index, OTU as column
cols=['Districts']+otus
abundanceTable = pd.DataFrame(index=samples,columns=cols)
print(abundanceTable)
```

```python
# generate abundance based on mean, for variance choose 1
for dist in meanMap.keys():
    for otu in meanMap[dist].keys():
        mu, sigma = meanMap[dist][otu],1
        s = np.random.normal(mu, sigma, count[dist])
        for i in range(len(s)):
            abundanceTable.loc[sampleDist[dist][i],otu]=s[i]
            abundanceTable.loc[sampleDist[dist][i],'Districts']=dist
writer = pd.ExcelWriter('rawRandomData.xlsx')
abundanceTable.to_excel(writer,'Sheet1')
```

```python
from scipy import stats
```

```python
res={}
abundanceTable=pd.read_excel('output.xlsx')
for column in abundanceTable.columns[1:]:
    df1=abundanceTable[column]
    df2=abundanceTable['Districts']
    df=df2.to_frame().join(df1)
    otuGroup=df.groupby('Districts')[column].apply(list)
    rows=[]
    for row in otuGroup:
        rows.append(row)
    res[column]=stats.kruskal(*rows).statistic
```

```python
outKruskalTest= pd.DataFrame(res,index=['Test-Statistic']).T
outKruskalTest.index=outKruskalTest.index.map(lambda x:int(x))
print(outKruskalTest.sort_index())
```
